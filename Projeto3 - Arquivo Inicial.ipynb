# Automação Web e Busca de Informações com Python

#### Desafio: 

Trabalhamos em uma importadora e compramos e vendemos commodities:
- Soja, Milho, Trigo, Petróleo, etc.

Precisamos pegar na internet, de forma automática, a cotação de todas as commodites e ver se ela está abaixo do nosso preço ideal de compra. Se tiver, precisamos marcar como uma ação de compra para a equipe de operações.

Base de Dados: https://drive.google.com/drive/folders/1O9aE_Hen5smf_a6TsbVF6DM0ms4Ykj6s?usp=share_link

Para isso, vamos criar uma automação web:

- Usaremos o selenium
- Importante: baixar o webdriver

```python
# !pip install selenium
```

```python
from selenium import webdriver


# passo 1 - entrar no navegador
navegador = webdriver.Chrome()
navegador.get('https://google.com')
```

```python
#passo 2 - importar a base de dados
import pandas as pd

tabela = pd.read_excel('commodities.xlsx')
display(tabela)
```

```python
#passo 3 - pegar a cotação dos produtos

for linha in tabela.index:
    produto = tabela.loc[linha,'Produto']
    produto = produto.replace('ó', 'o').replace('ã', 'a').replace('ú', 'u').replace('á', 'a').replace('ç', 'c').replace('é', 'e')
    navegador.get(f'https://www.melhorcambio.com/{produto}-hoje')
    valor = navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
    valor = valor.replace('.', '').replace(',', '.')
    valor = float(valor)
#passo 4 - colocar os dados na tabela
    tabela.loc[linha, 'Preço Atual'] = valor
    if tabela.loc[linha, 'Preço Ideal'] >= tabela.loc[linha, 'Preço Atual']:
        tabela.loc[linha, 'Comprar'] = 'COMPRAR'
    else:
        tabela.loc[linha, 'Comprar'] = 'VENDER'        
display(tabela)
```

```python
#passo 5 - sair do navegador e mandar os dados para a planilha
navegador.quit()

tabela = tabela.to_excel('commodities-alterado.xlsx', index=False)
```
